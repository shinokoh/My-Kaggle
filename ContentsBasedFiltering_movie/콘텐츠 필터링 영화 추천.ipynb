### 딕셔너리 형태의 컬럼 => 문자열 데이터를 가진 리스트로 변환
- genres
- keywords

```python
pd.set_option('max_colwidth', 100) #판다스 옵션 설정 함수:한 컬럼의 너비('max_colwidth')
movies_df[['genres', 'keywords']][:1]
```

```python
from ast import literal_eval # 문자열을 리스트 또는 딕셔너리 형태로 바꿔주는 함수

movies_df['genres'] = movies_df['genres'].apply(literal_eval)
movies_df['keywords'] = movies_df['keywords'].apply(literal_eval)
```

```python
# 'id', 'name' 중 문자열인 name키에 해당하는 값만 추출 
movies_df['genres'] = movies_df['genres'].apply(lambda x : [y['name'] for y in x])
movies_df['keywords'] = movies_df['keywords'].apply(lambda x : [y['name'] for y in x])
```

### genres 유사도 측정
- genres 문자열로 변경
- 문자열로 변환된 컬럼은 count 기반으로 피처 벡터화 변환(CountVectorizer) 
- genres 문자열을 피처 벡터화한 행렬로 변환한 데이터 세트를 '코사인 유사도'를 통해 비교, 데이터 세트의 레코드별로 타 레코드와 장르에서 유사도 값을 가지는 객체 생성
- 장르 유사도가 높은 영화 중 평점이 높은 순으로 영화 추천

```python
from sklearn.feature_extraction.text import CountVectorizer 

#CountVectorizer를 적용하기 위해 공백문자로 word 단위가 구분되는 문자열로 변환
movies_df['genres_literal'] = movies_df['genres'].apply(lambda x : (' ').join(x))
count_vect = CountVectorizer(min_df = 0, ngram_range=(1,2))
genre_mat = count_vect.fit_transform(movies_df['genres_literal'])
print(genre_mat.shape)
```

```python
# 코사인 유사도 반환 및 추출

from sklearn.metrics.pairwise import cosine_similarity

genre_sim = cosine_similarity(genre_mat, genre_mat)
print(genre_sim.shape)
print(genre_sim[:1])
```

```python
# 기준 행별로 비교 대상이 되는 행의 유사도 값이 높은 순으로 정렬된 행렬의 위치 인덱스 값을 추출

genre_sim_sorted_ind = genre_sim.argsort()[:, ::-1] #::-1 : 내림차순 정렬
print(genre_sim_sorted_ind[:1])
```

### 장르 콘텐츠 필터링을 이용한 영화 추천
- 장르 유사도에 따른 영화 추천 함수 생성

```python
def find_sim_movie(df, sorted_ind, title_name, top_n = 10):
    # 인자로 입력된 movies_df 데이터 프레임에서 'title' 컬럼이 입력된 title_name의 값인 데이터프레임을 추출
    title_movie = df[df['title'] == title_name]
    
    # title_name 을 가진 데이터 프레임의 index 객체를 ndarray로 반환하고
    # sorted_ind 인자로 입력된 genre_sim_sorted_ind 객체에서 유사도 순으로 top_n개의 index를 추출
    title_index = title_movie.index.values
    similar_indexes = sorted_ind[title_index, :(top_n)]
    
    # 추출된 top_n index 를 출력, top_n index는 2차원 데이터임
    # 데이터 프레임에서 인덱스로 사용하기 위해 1차원 데이터로 변경
    print(similar_indexes)
    similar_indexes = similar_indexes.reshape(-1)
    
    return df.iloc[similar_indexes]
```

```python
similar_movies = find_sim_movie(movies_df, genre_sim_sorted_ind, 'The Godfather', 10)
similar_movies[['title','vote_average']]
```

- 장르 유사도로 영화를 추천한 결과, 장르가 비슷한 영화도 알맞게 추천되었지만 평점이 0.0으로 낮은 영화도 추천됨
- 좀 더 많은 후보군을 선정하고 영화의 평점에 따라 필터링해 최종 추천하는 방식으로 변경할 필요가 있음

```python
movies_df[['title', 'vote_average', 'vote_count']].sort_values('vote_average', ascending = False)[:10]
```

- 소수의 관객이 특정 영화에 만점이나 매우 높은 평점을 부여해 왜곡된 데이터를 가지고 있는 경우가 있음
- 왜곡된 평점 데이터를 회피할 수 있도록 평가 횟수를 반영할 수 있는 새로운 평가 방식의 필요

#### 가중 평점 = (v/(v+m) * R + (m/v+m)) * c
- v : 개별 영화에 평점을 투표한 횟수
- m : 평점을 부여하기 위한 최소 투표 횟수 (가중치를 조절하는 역할 --> m 값을 높이면 투표횟수가 많은 영화에 더 많은 가중 평점을 부여)
- R : 개별 영화에 대한 평균 평점
- C : 전체 영화에 대한 평균 평점

```python
# m값은 전체 투표 횟수에서 상위 60%에 해당하는 횟수를 기준으로 정함

c = movies_df['vote_average'].mean() # 전체 평점 평균
m = movies_df['vote_count'].quantile(0.6) # 주어진 데이터를 분할하는 지점
print('c(전체 평점 평균) =',round(c,3), ', m(전체 투표수 상위 60%) =',round(m,3))
```

### 가중 평점을 적용하는 함수 생성

```python
percentile = 0.6
m = movies['vote_count'].quantile(percentile)
c = movies['vote_average'].mean()

def weighted_vote_avg(record):
    v = record['vote_count']
    R = record['vote_average']
    
    return ( (v/(v+m)) * R ) + (m/(m+v) + c )

movies_df['weighted_vote'] = movies_df.apply(weighted_vote_avg, axis = 1 )
```

```python
movies_df[['title', 'vote_average','weighted_vote','vote_count']].sort_values('weighted_vote', ascending = False)[:10]
```

- 이제 새롭게 정의된 평점 평균에 따라 장르 유사성이 높은 영화를 추천할 계획
- 장르 유사성이 높은 영화를 top_n의 2배수만큼 후보군을 선정해두고 weighted_vote 컬럼 값이 높은 순으로 top_n만큼 추출하는 방식

```python
def find_sim_movie(df, sorted_ind, title_name, top_n = 10):
    title_movie = df[df['title'] == title_name]
    title_index = title_movie.index.values
    
    # top_n의 2배에 해당하는 장르 유사성이 높은 인덱스 추출
    similar_indexes = sorted_ind[title_index, :(top_n*2)]
    print(similar_indexes)
    similar_indexes = similar_indexes.reshape(-1)
    
    #top_n의 2배에 해당하는 후보군에서 weighted_vote가 높은 순으로 top_n 만큼 추출
    return df.iloc[similar_indexes].sort_values('weighted_vote', ascending = False)[:top_n]

similar_movies = find_sim_movie(movies_df, genre_sim_sorted_ind, 'The Godfather', 10)
similar_movies[['title','vote_average','weighted_vote']]
```
